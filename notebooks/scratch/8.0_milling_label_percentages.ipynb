```python
import numpy as np
import pandas as pd
from pathlib import Path
import random
import matplotlib.pyplot as plt
import seaborn as sns

%load_ext autoreload
%autoreload 2
```

```python
path_processed_dir = Path(Path.cwd().parent.parent / 'data/processed/milling')
print(path_processed_dir)
```

```python
# load feature df
df = pd.read_csv(path_processed_dir / 'milling_features_comp_stride64_len1024' / 'milling_features_comp_stride64_len1024.csv')
df.head()
```

```python
n_sub_cuts = df.cut_id.unique().shape[0]
print("n_sub_cuts:", n_sub_cuts)
```

```python
# get the percentage of each tool_class
df_p = df.groupby('tool_class').size() / df.shape[0] * 100
df_p = df_p.reset_index()
df_p.columns = ['tool_class', 'percentage']

# get the count of each tool_class
df_c = df.groupby('tool_class').size().to_frame().reset_index()
df_c.columns = ['tool_class', 'count']

# merge the two dataframes
df_pc = df_p.merge(df_c, on='tool_class')[['tool_class', 'count', 'percentage']]

# show the percentage with only two decimal places
df_pc['percentage'] = df_pc['percentage'].round(2)
```
